```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load the merged data
merged_df = pd.read_csv('Task 3_Final_Content_Data_set.csv')

# Set a theme for the plots
sns.set_theme(style="whitegrid")

# Scenario 1: Top 5 Content Categories by Reactions (Bar Chart)
category_reaction_counts = merged_df.groupby('Category').size().reset_index(name='Reaction Count')
top_5_categories = category_reaction_counts.sort_values(by='Reaction Count', ascending=False).head(5)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_5_categories, x='Category', y='Reaction Count', palette='coolwarm')
plt.title('Top 5 Content Categories by Reactions', fontsize=16)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Reaction Count', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.savefig('top_5_categories.png')
plt.show()

# Scenario 2: Number of Unique Categories (Text Output)
unique_categories_count = merged_df['Category'].nunique()
print(f'Number of Unique Categories: {unique_categories_count}')

# Scenario 3: Number of Reactions to the Most Popular Category (Pie Chart)
most_popular_category = top_5_categories.iloc[0]
labels = [most_popular_category['Category'], 'Other Categories']
sizes = [most_popular_category['Reaction Count'], category_reaction_counts['Reaction Count'].sum() - most_popular_category['Reaction Count']]
colors = sns.color_palette('coolwarm', 2)
explode = (0.1, 0)

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140, textprops={'fontsize': 14})
plt.title('Reactions to the Most Popular Category vs Other Categories', fontsize=16)
plt.savefig('most_popular_category.png')
plt.show()


```
